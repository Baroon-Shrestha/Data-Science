```python
import pandas as pd
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
group_by_fare = df[["Sex","Fare"]].groupby(by = "Sex")
group_by_fare.mean()
```

```python
group_by_emb = df[["Embarked","Fare"]].groupby(by = "Embarked")
group_by_emb.mean()
```

```python
group_by_age = df[["Embarked","Age"]].groupby(by = "Embarked")
group_by_age.mean()
```

## Correlation

correlation varies from -1 - +1. The correlation shows the relation between two variables i.e. x and y.

*  **positive correlation** : when the value of x changes accordingly with the value of y.
*  **neagtive correlation** : when the value of x changes in opposite way with the value of y.
*  **No correlation** : when both x and y are independent variables and doesnot have any relation with each other.

```python
df[["Fare","Age"]].corr()
```
