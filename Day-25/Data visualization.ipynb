## Bar Graph

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df_sample = df.sample(frac=0.1)  # 10% of the original dataset
barG = df_sample.Survived.value_counts()

plt.bar(["Dead","Survived"],barG.to_numpy())

plt.show()
```
